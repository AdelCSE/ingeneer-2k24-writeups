### Import necessary libraries

````python
import matplotlib.pyplot as plt, numpy as np
from scipy.fftpack import dct
from PIL import Image
````

### Loading images

````python
original_img = Image.open('./images/original_rick.png')
modified_img = Image.open('./images/hidden_rick.png')
````

````python
fig, axs = plt.subplots(1, 2, figsize=(20, 8))
axs[0].imshow(original_img)
axs[0].set_title('Original Image')
axs[1].imshow(modified_img)
axs[1].set_title('Modified Image')
plt.show()
````

### Normalizing images

````python
norm_original = (np.array(original_img).astype(float) / 256)[:,:,:3]
norm_modified = (np.array(modified_img).astype(float) / 256)[:,:,:3]
````

### Difference between images

````python
difference = norm_modified - norm_original
````

### Discrete cosine transform (DCT)

````python
dct_img = dct(difference, axis=1, norm='backward')
plt.imshow(dct_img)
plt.show()
````

````python
flag = dct_img[85:115, 0:400]
plt.figure(figsize=(20,20))
plt.imshow( flag )
plt.show()
````

````python
transp = np.zeros_like(norm_original)
transp[85:115, 0:400] = flag[:, 0:400]

img = norm_original + transp

plt.imshow(img)
plt.show()
````
