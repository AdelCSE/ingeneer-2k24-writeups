### Import Necessary Libraries

````python
import umap
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
````

````python
import warnings
warnings.filterwarnings('ignore')
````

### Challenge creation

````python
# Generate some random data
data1, y1 = make_blobs(n_samples=1000, n_features=20, centers=20, shuffle=True, random_state=44)
data2, y2 = make_blobs(n_samples=1000, n_features=20, centers=16, shuffle=True, random_state=21)
data3, y3 = make_blobs(n_samples=1000, n_features=20, centers=12, shuffle=True, random_state=44)
data4, y4 = make_blobs(n_samples=1000, n_features=20, centers=18, shuffle=True, random_state=24)

data = [data1, data2, data3, data4]
y = [y1, y2, y3, y4]

# Apply dimensionality reduction (PCA) to transform data to 2D
plt.figure(figsize=(24, 4))
for i in range(1, 5):
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(data[i-1])
    plt.subplot(1,4,i)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y[i-1], cmap='viridis', edgecolor='k', alpha=0.8)
    plt.title(f'clusters{i}')
    plt.colorbar(label='Cluster')

# Plot the transformed data
plt.show()
````

### Save assets

````python
for i in range(1, 5):
    np.save(f'./data/clusters{i}.npy', data[i-1])
````

### Solution

````python
plt.figure(figsize=(20, 4))
for i in range(1, 5):
    x = np.load(f'./data/clusters{i}.npy')
    model = umap.UMAP()
    #model = TSNE() // that also works
    embs = model.fit_transform(x)
    plt.subplot(1, 4, i)
    plt.scatter(embs[:,0], embs[:,1])
    plt.title(f'clusters{i}')
    
plt.show()
````
